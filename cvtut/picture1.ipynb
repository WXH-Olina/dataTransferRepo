```python
import os
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def showImageOnNotebook(title, image):
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

openCV:BGR

matplotlib:RGB

```python
script_dir = os.getcwd()
image_path = os.path.join(script_dir, 'picture1.jpg')
img = cv.imread(image_path)
showImageOnNotebook('img', img)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
showImageOnNotebook('img_rgb', img_rgb)
```

```python
img_hsv = cv.cvtColor(img, cv.COLOR_RGB2HSV)
showImageOnNotebook('img_hsv', img_hsv)
```

```python
img_yuv = cv.cvtColor(img, cv.COLOR_BGR2YUV)
showImageOnNotebook('img_yuv', img_yuv)
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
showImageOnNotebook('img_gray', img_gray)
```
