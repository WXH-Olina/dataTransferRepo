```python
import os
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def showImageOnNotebook(title, image):
    plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# Load the image
script_dir = os.getcwd()
image_path = os.path.join(script_dir, "picture2.jpg")  # Replace with your image path
image = cv.imread(image_path)
showImageOnNotebook("original_image", image)
```

Method 1:

```python
# Convert the image to HSV color space
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
showImageOnNotebook("gray_image", gray_image)
```

```python
ret, target_image = cv.threshold(gray_image, 215, 255, cv.THRESH_BINARY)
showImageOnNotebook("target_image", target_image)
```

Method 2:

```python
# Convert the image to HSV color space
hsv_image = cv.cvtColor(image, cv.COLOR_BGR2HSV)

# Define the thresholds for gold/yellow color
lower_yellow = np.array([20, 100, 100])  # Lower bound for yellow
upper_yellow = np.array([30, 255, 255])   # Upper bound for yellow

# Create a mask for the yellow color
mask_yellow = cv.inRange(hsv_image, lower_yellow, upper_yellow)
showImageOnNotebook("mask_yellow", mask_yellow)
```

```python
ret, binary_image = cv.threshold(mask_yellow, 1, 255, cv.THRESH_BINARY)
showImageOnNotebook("binary_image", binary_image)
```
